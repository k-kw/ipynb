```python
#MNIST用意
import tensorflow as tf
print(tf.__version__)

import keras as ks
print(ks.__version__)

import os 
os.environ['KERAS_BACKEND']='theano'

from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
from keras.utils import np_utils

from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
mnist=fetch_openml('mnist_784')

x=mnist['data']
x=x.to_numpy()
x=x.tolist()

print('データ取り込み完了')


from PIL import Image
dat_file_name='mnist_expand.dat'
N=70000
width=28
height=28
ex_width=128
ex_height=128
OUTPUT_FOLDER='./output/'
if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)

f=open(dat_file_name,'wb')

for k in range(N):

    #画像作成
    image_buffer=Image.new('RGB',(width,height),(255,255,255))

    for h in range(height):
        for w in range(width):           
            pixel=int(x[k][h*width+w])
            image_buffer.putpixel((w,h),(pixel,pixel,pixel))
    image_buffer.save(OUTPUT_FOLDER+'tmp.jpg','JPEG')


    #画像拡大
    im=Image.open(OUTPUT_FOLDER+'tmp.jpg')
    expand_img=im.resize(size=(ex_width,ex_height))
    expand_img.save(OUTPUT_FOLDER+'tmp_expand.jpg','JPEG')
    im.close()
    

    #拡大後画像データ取得、画像削除
    im2=Image.open(OUTPUT_FOLDER+'tmp_expand.jpg')
    ex_x_tup=[]
    for h in range(ex_height):
        for w in range(ex_width):
            tmp=im2.getpixel((w,h))
            ex_x_tup.append(tmp)
    im2.close()
    
    #print(ex_x_tup)

    #print('\n')
    #print('\n')

    #getpixelがタプルなのでintとして取り出す
    ex_x=[]
    for h in range(ex_height):
        for w in range(ex_width):
            tmp=ex_x_tup[h*ex_width+w][0]
            ex_x.append(int(tmp))

    #os.remove(OUTPUT_FOLDER+'tmp_'+str(k)+'_'+str(ex_width)+'x'+str(ex_height)+'.jpg')
    #os.remove(OUTPUT_FOLDER+'tmp'+str(k)+'.jpg')
    del ex_x_tup

    #print(ex_x)

    #書き込み
    for i in range(len(ex_x)):
        tmp=ex_x[i]
        tmp=tmp.to_bytes(4,'little')
        f.write(tmp)
    del ex_x
    
    if((k+1)%10000==0):
        print(str(k)+'まで完了')
    
    
    if(k==N-1):
        print('終了')
    
    
    



f.close()
```
