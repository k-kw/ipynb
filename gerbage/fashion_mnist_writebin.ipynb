```python
def label_wright(dp,labels):
  with open(dp,'wb') as f:
    for label in labels:
      tmp=int(label)
      tmp=tmp.to_bytes(4,'little')
      f.write(tmp)
```

```python
label_wright('./mnist_label.dat',ym)
label_wright('./fm_label.dat',yf)
```

```python
def label_read(dp,num):
  labellist=[]
  with open(dp,'rb') as f:
    for i in range(num):
      tmp=f.read(4)
      q=int.from_bytes(tmp,'little')
      labellist.append(q)
  return labellist
```

```python
label_mnist=label_read('./mnist_label.dat',70000)
label_fm=label_read('./fm_label.dat',70000)
```
