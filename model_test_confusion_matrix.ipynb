```python
SX=128
SY=128
N=70000
dataname='mnist_diff_expand_128_0.2_0.1_sim.dat'
modelname='mnist_diff_expand_128_0.2_0.1_epochs50.h5'
matrixname='mnist_line_mat_diff_expand_128_0.2_0.1_epochs50.png'
resoluion=2**8
x=[]
with open(dataname,'rb') as f:
         for i in range(N*SX):
             tmp=f.read(4)
             q=int.from_bytes(tmp,'little')
             x.append(q)

import numpy as np
x=np.array(x)
x=x.reshape(N,SX)
import keras as ks
model=ks.models.load_model(modelname)

import tensorflow as tf

from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784')


import pandas as pd

y=mnist['target']
y=y.to_numpy()

x_test=x[60000:]
y_test=y[60000:]


x_test=x_test.astype('float32')

x_test/=(resoluion-1)

#混同行列
test_predictions = model.predict(x_test)
test_predictions=np.argmax(test_predictions, axis=1)

y_test=y_test.astype('int64')

print(type(y_test))
print(type(x_test))
print(type(test_predictions))
print(y_test.shape)
print(x_test.shape)
print(test_predictions.shape)

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
_, ax = plt.subplots(figsize = (10, 10))

cm = confusion_matrix(y_true = y_test, y_pred = test_predictions)
sns.heatmap(cm,square=True,cmap='Blues',annot=True,fmt='d',ax=ax,vmax=50,vmin=0)
plt.savefig(matrixname)
```
